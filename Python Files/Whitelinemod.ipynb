```python
import requests
import xml.etree.ElementTree as ET

# Send a request to the XML file
response = requests.get("https://www.whitelinemod.com/wp-sitemap-posts-product-1.xml")

# Parse the XML content
xml_content = response.content.decode("utf-8")
xml_tree = ET.fromstring(xml_content)

# Find all 'loc' tags
link_tags = xml_tree.findall(".//{http://www.sitemaps.org/schemas/sitemap/0.9}loc")

# Extract the links and put them in a list
links = [tag.text for tag in link_tags]
```

```python
products = {'title': [], 'description': [], 'color' : [] , 'sku' : [] , 'category' : [] , 'desc_2' : [] , 'add' : []}

for link in links:
    url = link
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    title_div = soup.find('div', {'class': 'summary entry-summary'})
    desc_div = soup.find('div', {'class': 'woocommerce-product-details__short-description'})
    color_div = soup.find('div', {'class': 'select swatch-control'})
    sku_div = soup.find('span', {'class': 'sku'})

    desc_div_1 = soup.find('div', {'class': 'woocommerce-Tabs-panel woocommerce-Tabs-panel--description panel entry-content wc-tab'})
    soup_2 = BeautifulSoup(str(desc_div_1), 'html.parser')
    description_div = soup.find('div', {'id': 'tab-description'})
    add_span = soup.find('div', {'class': 'woocommerce-Tabs-panel woocommerce-Tabs-panel--additional_information panel entry-content wc-tab'})


# append text inside h2 tags to description_text
    description_text = ''
    if description_div is not None:
        h2_tags = description_div.find_all('h2')
        for h2 in h2_tags:
            description_text += h2.text + '\n'

        # append text inside p tags to description_text
        p_tags = description_div.find_all('p')
        for p in p_tags:
            description_text += p.text + '\n'

        # append text inside li tags to description_text
        li_tags = description_div.find_all('li')
        for li in li_tags:
            description_text += li.text + '\n'


    if add_span is not None:
        table_rows = add_span.find_all('tr')

    table_data = {}

    if table_rows is not None:
      for row in table_rows:
          cells = row.find_all(['th', 'td'])
          key = cells[0].text.strip()
          value = cells[1].text.strip()
          table_data[key] = value



    if desc_div is not None:
        products['description'].append([p_tag.text.strip() for p_tag in desc_div.find_all('p')])
    else:
        products['description'].append([])


    if color_div is not None:
      products['color'].append(list(filter(lambda x: x != '', [option['value'] for option in soup.find('select').find_all('option')])))

    else:
        products['color'].append([])

    products['title'].append(title_div.find('h1').text.strip())
    products['sku'].append(soup.find('span', {'class': 'sku'}).text)
    products['category'].append(soup.find('span', {'class': 'posted_in'}).text)
    products['desc_2'].append(description_text)
    products['add'].append(table_data)

import pandas as pd
df = pd.DataFrame(products)

df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import json

def extract_data(link):
    data = {}
    try:
        response = requests.get(link)
        soup = BeautifulSoup(response.content, 'html.parser')
        title_div = soup.find('div', {'class': 'summary entry-summary'})
        desc_div = soup.find('div', {'class': 'woocommerce-product-details__short-description'})
        color_div = soup.find('div', {'class': 'select swatch-control'})
        sku_div = soup.find('span', {'class': 'sku'})

        desc_div_1 = soup.find('div', {'class': 'woocommerce-Tabs-panel woocommerce-Tabs-panel--description panel entry-content wc-tab'})
        soup_2 = BeautifulSoup(str(desc_div_1), 'html.parser')
        description_div = soup.find('div', {'id': 'tab-description'})
        add_span = soup.find('div', {'class': 'woocommerce-Tabs-panel woocommerce-Tabs-panel--additional_information panel entry-content wc-tab'})

        data['title'] = title_div.find('h1').text.strip()
        data['sku'] = sku_div.text.strip()
        data['category'] = soup.find('span', {'class': 'posted_in'}).text.strip()

        description_text = ''
        if description_div is not None:
            description_text = '\n'.join([h2.text + '\n' + p.text for h2 in description_div.find_all('h2') for p in h2.find_next_siblings('p')])
            description_text += '\n'.join([li.text for li in description_div.find_all('li')])

        data['description'] = [p_tag.text.strip() for p_tag in desc_div.find_all('p')] if desc_div is not None else []

        data['color'] = list(filter(lambda x: x != '', [option['value'] for option in soup.find('select').find_all('option')])) if color_div is not None else []

        table_data = {}
        if add_span is not None:
            table_rows = add_span.find_all('tr')
            for row in table_rows:
                try:
                    key = row.find('th').text.strip()
                    value = row.find('td').text.strip()
                    table_data[key] = value
                except:
                    pass

        data['add'] = table_data
        data['desc_2'] = description_text
    except:
        pass

    return data

products = [extract_data(link) for link in links]

with open('products.json', 'w') as f:
    json.dump(products, f)
```
