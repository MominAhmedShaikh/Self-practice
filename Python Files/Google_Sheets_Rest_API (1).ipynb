```
function doGet(req) {
  var sheet = SpreadsheetApp.getActiveSpreadsheet().getSheetByName("Review");
  var data = sheet.getDataRange().getValues(); // Get all the data in the sheet
  var headers = data.shift(); // Remove the headers from the data and store them separately

  var json = [];

  for (var i = 0; i < data.length; i++) {
    var obj = {};
    for (var j = 0; j < headers.length; j++) {
      obj[headers[j]] = data[i][j];
    }
    json.push(obj);
  }
  return  ContentService.createTextOutput(JSON.stringify({data : json})).setMimeType(ContentService.MimeType.JSON);

}

```
